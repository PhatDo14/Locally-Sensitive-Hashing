```python
# khởi tạo data
a = "flying fish flew by the space station"
b = "we will not allow you to bring your pet armadillo along"
c = "he figured a few sticks of dynamite were easier than a fishing pole to catch fish"
```

```python
# 1st step : Shingling data set
def shingle(text: str, k: int):                      # create a shingle function gồm : chuỗi và độ dài k-shingle
    shingle_set = []
    for i in range(len(text) - k+1):
        shingle_set.append(text[i:i+k])               # Lấy các chuỗi text có độ dài k vào vào 1 mảng tạo thành 1 shingle set
    return set(shingle_set)
```

```python
k=2
a = shingle(a, k)
b = shingle(b, k)
c = shingle(c, k)
print(a)
```

```python
# tiếp theo tạo thành 1 vocab  bằng cách kết hợp (union) tất cả các data ban đầu thành 1 data lớn

vocab=list(a.union(b).union(c))
vocab
```

```python
# tạo 1 one hot
# Ta sẽ chạy vòng loop qua a và kiếm trong vocab với từng cái shingle mà nếu có tồn tại trong signature thì cho nó = 1 còn không thì = 0
a_1hot = [1 if x in  a else 0 for  x in vocab]
b_1hot = [1 if x in  b else 0 for  x in vocab]
c_1hot = [1 if x in  c else 0 for  x in vocab]
```

```python
a_1hot
```

```python
# min hashing

# tạo 1 mảng hoán vị ngẫu nhiên mà có chỉ số từ 1 cho đến độ dài của vocab và sau đó suffle và tìm kiếm cho đến khi mình tìm được số 1

hash_ex = list(range(1, len(vocab)+1))    # tạo 1 list từ 1 đến độ dài của vocab
print(hash_ex)  # we haven't shuffled yet
```

```python
from random import shuffle

shuffle(hash_ex)
print(hash_ex)
```

```python
for i in range(1,10):
  print(f"{i} -> {hash_ex.index(i)}")
```

```python
for i in range(1, len(vocab)+1):
    idx = hash_ex.index(i)
    signature_val = a_1hot[idx]
    print(f"{i} -> {idx} -> {signature_val}")
    if signature_val == 1:
        print('match!')
        break
```

```python
# tóm tắt lại các ý trên

def create_hash_func(size: int):
    # function for creating the hash vector/function
    hash_ex = list(range(1, len(vocab)+1))
    shuffle(hash_ex)
    return hash_ex

def build_minhash_func(vocab_size: int, nbits: int):
    # function for building multiple minhash vectors
    hashes = []
    for _ in range(nbits):
        hashes.append(create_hash_func(vocab_size))
    return hashes

# we create 20 minhash vectors
minhash_func = build_minhash_func(len(vocab), 20)
```

```python
def create_hash(vector: list):
    # use this function for creating our signatures (eg the matching)
    signature = []
    for func in minhash_func:
        for i in range(1, len(vocab)+1):
            idx = func.index(i)
            signature_val = vector[idx]
            if signature_val == 1:
                signature.append(idx)
                break
    return signature
```

```python
# now create signatures
a_sig = create_hash(a_1hot)
b_sig = create_hash(b_1hot)
c_sig = create_hash(c_1hot)

print(a_sig)
print(b_sig)
```

Tính độ tương đồng Jaccard

```python
# Kiểm tra lại độ giống nhau giũa 2 data
def jaccard(a: set, b: set):
    return len(a.intersection(b)) / len(a.union(b))
```

```python
jaccard(a, b), jaccard(set(a_sig), set(b_sig))
```

```python
jaccard(a, c), jaccard(set(a_sig), set(c_sig))
```

```python
jaccard(b, c), jaccard(set(b_sig), set(c_sig))
```

```python
# final process LSH
def split_vector(signature, b):
    assert len(signature) % b == 0
    r = int(len(signature) / b)
    # code splitting signature in b parts
    subvecs = []
    for i in range(0, len(signature), r):
        subvecs.append(signature[i : i+r])
    return subvecs
```

```python
band_a = split_vector(a_sig, 10)
band_b = split_vector(b_sig, 10)
band_b
```

```python
band_c = split_vector(c_sig, 10)
band_c
```

Sau đó, ta lặp qua các danh sách để xác định xem có bất kỳ sự trùng khớp nào giữa các vectơ con không. Nếu tìm thấy bất kỳ kết quả phù hợp nào -> sẽ coi các vectơ đó là cặp ứng cử viên.

**MinHash Example**

```python
# Import the required libraries
from datasketch import MinHash, MinHashLSH
```

Threshold=0.5

```python
# Create a MinHash LSH index
lsh = MinHashLSH(threshold=0.5, num_perm=128)

# Create MinHash signatures for documents
doc1 = "This is the first document"
doc2 = "This document is the second document"
doc3 = "And this is the third one"

minhash1 = MinHash()
minhash2 = MinHash()
minhash3 = MinHash()

for word in doc1.split():
  minhash1.update(word.encode('utf8'))

for word in doc2.split():
  minhash2.update(word.encode('utf8'))

for word in doc3.split():
  minhash3.update(word.encode('utf8'))

# Index the documents
lsh.insert("doc1", minhash1)
lsh.insert("doc2", minhash2)
lsh.insert("doc3", minhash3)

# Query similar documents
query_doc = "This is a document"
query_minhash = MinHash()
for word in query_doc.split():
  query_minhash.update(word.encode('utf8'))
result = lsh.query(query_minhash)
print("Similar documents:", result)
```
